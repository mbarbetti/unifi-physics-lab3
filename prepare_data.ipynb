```python
import os
import numpy as np
import pandas as pd

from datetime import datetime
from utils import data_processing
```

```python
files_path = "../Gruppi/Gruppi"
export_path = "data/2021"

df = data_processing (files_path, export_path, tdc_resolution = 1e-9)
df
```

```python
timestamp = str (datetime.now()) . split (".") [0]
timestamp = timestamp . replace (" ","_")
version = ""
for time, unit in zip ( timestamp.split(":"), ["h","m","s"] ):
  version += time + unit   # YYYY-MM-DD_HHhMMmSSs

file_name = f"manipulation_{version}"

df.to_csv ( f"data/manipulation/{file_name}.txt", header = df.columns, index = None, sep = " " )
```
